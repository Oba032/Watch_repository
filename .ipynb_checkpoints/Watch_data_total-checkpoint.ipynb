```python
df = pd.read_csv('HealthAutoExport-2024-12-21-2025-03-21.csv', sep=';')
print(df.head())
```

```python
print(df.columns)
print(df.dtypes)
print(df.describe())
```

```python
grouped = df.groupby('Date').agg({
    'Active Energy (kcal)': 'sum',
    'Apple Exercise Time (min)': 'sum',
    'Apple Move Time (min)': 'sum',
    'Apple Stand Hour (hours)': 'sum',
    'Heart Rate [Min] (bpm)': 'mean',
    'Heart Rate [Max] (bpm)': 'mean',
    'Heart Rate [Avg] (bpm)': 'mean',
    'Heart Rate Variability (ms)': 'mean',
    'Respiratory Rate (count/min)': 'mean',
    'Resting Energy (kJ)': 'mean',
    'Resting Heart Rate (bpm)': 'mean',
    'Sleep Analysis [Asleep] (hr)': 'sum',
    'Sleep Analysis [In Bed] (hr)': 'sum',
    'Sleep Analysis [Core] (hr)': 'sum',
    'Sleep Analysis [Deep] (hr)': 'sum',
    'Sleep Analysis [REM] (hr)': 'sum',
    'Sleep Analysis [Awake] (hr)': 'sum',
    'Step Count (steps)': 'sum',
    'Walking Heart Rate Average (bpm)': 'mean',
    'Walking Speed (km/hr)': 'mean'
}).reset_index()

print(grouped)
```

```python
list_of_days = grouped.to_dict(orient='records')

print(list_of_days)
```

```python
kJ_to_kcal = 0.239006
```

```python
df['Resting Energy (kcal)'] = df['Resting Energy (kJ)'] * kJ_to_kcal
```

```python
grouped['Resting Energy (kcal)'] = grouped['Resting Energy (kcal)'].round(1)
print(grouped)
```

```python
grouped['Active Energy (kcal)'].round(1)
grouped['Resting Energy (kJ)'].round(1)
print(grouped)
```

```python
grouped['Heart Rate [Min] (bpm)'] = grouped['Heart Rate [Min] (bpm)'].round(0).astype(int)
grouped['Heart Rate [Max] (bpm)'] = grouped['Heart Rate [Max] (bpm)'].round(0).astype(int)
print (grouped)
```
